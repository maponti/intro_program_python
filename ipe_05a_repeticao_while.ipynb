## Laço `while` 

A estrutura `while` é mais geral do que a `for` e permite repetir um bloco de código mediante uma *condição*

Vamos relembrar a estrutura condicional `if` que tinha a seguinte sintaxe:

```
if <condicao>:
    <bloco indentado, executado se condicao for verdadeira>
    
<comando nao indentado, fora do if>
```

Nessa estrutura, caso a condição seja verdadeira, 
* o programa executa o `bloco indentado`
* a seguir, sai da estrutura `if` e executa o `comando nao indentado`
Caso a condição seja falsa
* pula o bloco `if` e executa diretamente o `comando nao indentado`

---

No caso de um `while` a sintaxe é 
```
while <condicao>:
    <bloco indentado, executado se condicao for verdadeira>
    
<comando nao indentado, fora do while>
```

Assim o `bloco indentado` pode ser executado *múltiplas* vezes, **enquanto** a condição for verdadeira

Vamos relembrar o problema de divisão por zero

````python
print("Operação de Divisão")
x = float(input("entre com numerador: "))
y = float(input("entre com denominador: "))

print(x/y)
````

Uma forma era não executar a divisão se o denominador for zero

````python
print("Operação de Divisão")
x = float(input("entre com numerador: "))
y = float(input("entre com denominador: "))

if y != 0:
    print(x/y)
````

Agora podemos repetir a entrada de dados enquanto o denominador for zero

````python
print("Operação de Divisão")
x = float(input("entre com numerador: "))

y = 0
while (y == 0):
    y = float(input("entre com denominador (não zero): "))

print(x/y)
````

No `while` é preciso definir a condição muito bem, para que haja uma **condição de parada**, do contrário pode-se entrar em `loop` infinito, o que ocorre quando a condição é sempre verdadeira


````python
while (True):
    print("infinito")
````

````python
N = int(input()) # usuario entra com número

i = 10

# enquanto i for maior do que N
while (i > N):
    print("loop", i) # imprime
    i = i + 1     # incrementa (aumenta em 1) o valor de i
````

#### <font color="blue">Exercício 5b.1 </font>

Faça um programa que receba como entrada um número inteiro, e divida esse número (usando divisão inteira) por 2 sucessivamente (imprimindo a cada divisão) até que o resultado seja 1. Ao final, mostra o número de divisões que foram necessárias para atingir a *condição de parada*  do while


OBS: note que essa contagem resulta no logaritmo na base 2, caso o número seja uma potência de 2

---


````python
num = int(input())
c = 0
while (num > 1):
    num = num//2
    c = c + 1
    print(num)
print("Divisoes por 2: ", c)
````

Uma das utilidades do while é manter um bloco de código funcionando enquanto o usuário deseja. Por exemplo podemos criar um programa calculadora que se repete enquanto o usuário não digitar a operação 0 para sair.

````python
# operacao inicializando
op = -1
while (op != 0):
    print("\nCalculadora")
    x = float(input("Valor 1: "))
    y = float(input("Valor 2: "))
    op = int(input("Digite 1 para soma, 2 para subtracao ou 0 para sair: "))
    if (op == 1):
        print("Resultado:", x+y)
    elif (op == 2):
        print("Resultado:", x-y)    
````

### Aninhando laços while

Assim como no caso do `for` também é possível aninhar `while`

Suponha que tenhamos uma lista de precos contendo frutas e preços que foram conseguidos em diferentes compras, e que desejamos:
* para cada lista, da primeira para a última: remova e imprima todos os preços do fim para o início, até que todos sejam removidos, 
* remover a lista após todos os preços serem removidos
* repetir até que a lista `precos` seja vazia

Para remover usamos o método `pop(indice)`

````python
precos = [[25.0, 27.0, 26, 24],
          [4.99, 4.5, 5.1],
          [9.99, 10, 9.5, 8.9, 11.00],
          [2.50, 2.5]]
````

````python
i = 0
while (len(precos) > 0):
    while (len(precos[0]) > 0):
        pr = precos[0].pop(-1)
        print(pr, end=', ')
    precos.pop(0)
    print()
    
print(precos)
````
